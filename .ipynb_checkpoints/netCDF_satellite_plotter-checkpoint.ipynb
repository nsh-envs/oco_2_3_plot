# NetCDF plotting of OCO2 and OCO3 data

```python
# Import packages
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# General definitions
pi = np.pi
inf = np.inf
NaN = np.NaN
```

## Examples from cartopy documentation

```python
lon = np.linspace(-80, 80, 25)
lat = np.linspace(30, 70, 25)
lon2d, lat2d = np.meshgrid(lon, lat)

data = np.cos(np.deg2rad(lat2d) * 4) + np.sin(np.deg2rad(lon2d) * 4)
```

```python
# The projection keyword determines how the plot will look
plt.figure(figsize=(6, 3))
ax = plt.axes(projection=ccrs.NorthPolarStereo(central_longitude=0))
# ax.set_global()
ax.coastlines()

ax.contourf(lon, lat, data, transform=ccrs.PlateCarree())  # didn't use transform, but looks ok...
plt.show()
```

```python
ax = plt.axes(projection=ccrs.Mollweide())
ax.stock_img()

ny_lon, ny_lat = -75, 43
delhi_lon, delhi_lat = 77.23, 28.61

plt.plot([ny_lon, delhi_lon], [ny_lat, delhi_lat],
         color='blue', linewidth=2, marker='o',
         transform=ccrs.Geodetic(),
         )

plt.plot([ny_lon, delhi_lon], [ny_lat, delhi_lat],
         color='gray', linestyle='--',
         transform=ccrs.PlateCarree(),
         )

plt.text(ny_lon - 3, ny_lat - 12, 'New York',
         horizontalalignment='right',
         transform=ccrs.Geodetic())

plt.text(delhi_lon + 3, delhi_lat - 12, 'Delhi',
         horizontalalignment='left',
         transform=ccrs.Geodetic())

plt.show()
```

```python
# Extra settings
day = 12
hour = 12

# Setting data path
data = '/home/niels/WinDoc/Data/'

# Select diretory
dir_num = 0
OCOX = ['OCO2/', # 0
        'OCO3/', # 1
       ][dir_num]

path = data + OCOX

filenames = np.loadtxt(path+"filenames.txt",dtype='str')

ds = xr.open_dataset(path+filenames[0], engine='netcdf4')
ds
```

```python
spec = ['xco2','latitude','longitude']
da = ds[spec] #.mean(dim='time')
min_values = np.zeros(3)
max_values = np.zeros(3)
#lat, lon = ds['latitude'].values, ds['longitude'].values
for i in range(len(spec)):
    min_values[i] = da[spec[i]].min().values
    max_values[i] = da[spec[i]].max().values
    print(f'{spec[i]:<10}',': {:6.2f} {:6.2f}'.format(min_values[i],max_values[i]))

print(72*'-')

# Either use sel/isel (selecting data from values/index)
# or use .where to do conditional selection.

# Create mask
# name  = [min, max]
val_lim = [-inf, inf] 
lat_lim = [-20,90]
lon_lim = [-90,90]

lat_mask = (da.latitude  >= lat_lim[0]) & (da.latitude  <= lat_lim[1])
lon_mask = (da.longitude >= lon_lim[0]) & (da.longitude <= lon_lim[1])
val_mask = (da[spec[0]]  >= val_lim[0]) & (da[spec[0]]  <= val_lim[1])

mask = lat_mask & lon_mask # & val_mask

# Select case
case = "none"

# sel/isel
if case == "sel":
    selelection = {spec[0]:slice(-inf,inf),
                   spec[1]:slice(lat_lim[0],lat_lim[1]),
                   spec[2]:slice(lon_lim[0], lon_lim[1]),
                  }
    plot_data = da.sel(selection)

# .where
if case == "where":
    plot_data = da.where(mask,drop=True) # drop=True -> reduce dimensions to only the variables used
    
# No selection
if case == "none":
    plot_data = da

# Construct numpy arrays
plot_val = plot_data[spec[0]].values
lat_val  = plot_data[spec[1]].values
lon_val  = plot_data[spec[2]].values

print(f'{spec[0]:<10}',f': size {plot_val.size}')
print(f'{spec[1]:<10}',f': size {lat_val.size}')
print(f'{spec[2]:<10}',f': size {lon_val.size}')
```

```python
## plot the first layer
# Cartopy uses to very important keywords to handle the different crs (Coordinate Reference System)
# projection: This is the map projection we want to plot *onto*. This set the crs for the axes and the plot.
#             It also influences the ax.coastlines and ax.gridlines functions.
# 

map_proj   = ccrs.Mollweide()
data_proj  = ccrs.PlateCarree()

fig = plt.figure(figsize=(12,8))
ax  = plt.axes(projection=map_proj)

ax.set_global()
ax.coastlines(color='slategray',
              zorder=2,
              alpha=0.5,
              linewidth=0.7,
             )

if ax.projection == ccrs.PlateCarree():
    gl = ax.gridlines(draw_labels=True,
                      zorder=1,
                      alpha=0.5,
                      linewidth=0.5,
                     )
if ax.projection == ccrs.Mollweide():
    gl = ax.gridlines(draw_labels=True,
                      zorder=1,
                      alpha=0.5,
                      linewidth=0.5,
                     )

    
    gl.xlabel_style = {'size':7}
    gl.ylabel_style = {'size':7}

cmap = plt.get_cmap('jet')
cmap.set_bad ('w',1.0)

# Need lon, lat in 2D
im = ax.scatter(lon_val, lat_val, c=plot_val,
                cmap=cmap,
                vmin=380,
                vmax=420,
                transform=data_proj,
               )

plt.colorbar(im, ax=ax, orientation='vertical',label='CO2 [ppm]')

plt.title(")

if False:
    ax.set_xlim(37,40)
    ax.set_ylim(22.5,25)
```
